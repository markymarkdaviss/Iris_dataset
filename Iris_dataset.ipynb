```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df
```

```python
df.Species.unique()
```

```python
df.isnull().sum()
```

```python
df['Species'] = np.where(df['Species'] == 'Iris-setosa', 0, np.where(df['Species'] == 'Iris-versicolor', 1, 2))
```

```python
df.drop('Id', axis=1)
```

## Visualizing the data

```python
# sns.boxplot(x=df['PetalLengthCm'], y=df['PetalWidthCm'], hue=df['Species'])
sns.pairplot(df, hue="Species")
```

## Putting data into algorithm

```python
from sklearn.model_selection import train_test_split
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
Y = df['Species']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state=52)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
linreg_predict = linreg.predict(X_test)
linreg_score = linreg.score(X_train, y_train)
linreg_score
```

## K nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(X_train, y_train)
KNN_predict = KNN.predict(X_test)
KNN_score = KNN.score(X_train, y_train)
KNN_score
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
linreg = LogisticRegression()
linreg.fit(X_train, y_train)
linreg_predict = linreg.predict(X_test)
linreg_score = linreg.score(X_train, y_train)
linreg_score
```
